```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

def show_wordcloud(data, title = None):
    wordcloud = WordCloud(
        background_color = 'white',
        max_words = 200,
        max_font_size = 40, 
        scale = 3,
        random_state = 42
    ).generate(str(data))

    fig = plt.figure(1, figsize = (20, 20))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize = 20)
        fig.subplots_adjust(top = 2.3)

    plt.imshow(wordcloud)
    plt.show()
    
# print wordcloud
show_wordcloud(train["cleaned review"])
```

# upsampling fake review cases

```python
import imblearn
```

```python

from imblearn.over_sampling import RandomOverSampler
```

```python
import numpy as np
train_y_arr = np.array(train_y['label'])
```

```python
# ros random over sampler
ros = RandomOverSampler()
train_X_ros, train_y_ros = ros.fit_sample(train_X, train_y_arr)
```

```python
train_X_ros = pd.DataFrame(train_X_ros, columns = ['ex_id','user_id','prod_id','rating','date','review','cleaned review','text length','word length'])
train_X_ros.head()
# not sure why but on my local machine, the trian_X_ros after oversampling is still DataFrame
# But on colab version it turned to array
```

```python
print('pre-upsampling genuine cases: ',train_y[train_y['label'] == 0].shape[0])
print('post-upsampling genuine cases: ',train_y_ros[train_y_ros == 0].shape[0])
print('pre-upsampling fake cases: ',train_y[train_y['label'] == 1].shape[0])
print('post-upsampling fake cases: ',train_y_ros[train_y_ros == 1].shape[0])
```

# bag of words

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
#BoW_data = count_vect.fit_transform(train_X['cleaned review'].values)
count_vect.fit(train_X_ros['cleaned review'])
```

```python
BoW_train_X = count_vect.transform(train_X_ros['cleaned review'])
```

```python
BoW_val_X = count_vect.transform(val_X['cleaned review'])
```

```python
BoW_train_X.shape, BoW_val_X.shape
```

# Naive Bayes

```python
from sklearn import naive_bayes
from sklearn.metrics import roc_curve, auc
```

```python
nb_bernoulli = naive_bayes.BernoulliNB()
nb_bernoulli.fit(BoW_train_X, train_y_ros)
```

```python
preds_nb = nb_bernoulli.predict_proba(BoW_val_X)[:,1]
```

```python
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC')

nb_fpr, nb_tpr, nb_thresholds = roc_curve(val_y, preds_nb)
nb_roc_auc = auc(nb_fpr, nb_tpr)
nb_curve = plt.plot(nb_fpr, nb_tpr, label = 'BernoulliNB' + ' (AUC = %0.3f)' % nb_roc_auc)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(val_y, nb_bernoulli.predict(BoW_val_X))
```
